```
embeding_index={}

f=open('glove.42B.300d.txt',encoding='utf-8')

for line in f:
  #reading one line
    values=line.split() #splitting the 1st line or string with spaces,will give list of len(301).1 word and 300 nos.
    word=values[0] # word (starting word of line) would be the 0th elemant
    coefs=np.asarray(values[1:],dtype='float32') #values[1:]=corresponding 300 dim vectors.from1 to 300
    embeding_index[word]=coefs
f.close()
# this again will take good amount of time
```

```
from sklearn.model_selection import train_test_split
data_train,data_test=train_test_split(labels,test_size=0.2)
```

```
x_train=data_train['question_text']
y_train=data_train['target']
x_test=data_test['question_text']
y_test=data_test['target']
```

```
from nltk import word_tokenize
```

```
import nltk #Natural Language Toolkit
nltk.download('punkt')
#q-punkt
```

```
sent_lens=[len(word_tokenize(x)) for x in x_train]
```

```
np.percentile(sent_lens,99)
```

```
max_len=44 #every sentence or every review wwill have only 44 words
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.layers import Flatten
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Lambda,Conv1D, MaxPooling1D,Activation, Dense,BatchNormalization,Input,Embedding,Dropout
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.layers import LSTM
```

```
tk=Tokenizer(char_level=False,split=' ') #spliting sentences with spaces

tk.fit_on_texts(x_train) #

seq_train=tk.texts_to_sequences(x_train) #getting unique words in text
seq_test=tk.texts_to_sequences(x_test) #unique words in text
```

```
x_train[924911]
```

```
seq_train[0]
```

```
vocab_size=len(tk.word_index)
```

```
vocab_size
```

```
#tk.word_index #index dictionary with every word
```

```
seq_train_matrix=sequence.pad_sequences(seq_train,maxlen=max_len) #padding/truncating sentences/reviews tomake it of size 44
seq_test_matrix=sequence.pad_sequences(seq_test,maxlen=max_len)
```

```
seq_train_matrix.shape
```

```
seq_train_matrix
```

```
# creating our own embedding matrix to bring down the size to 300
# we'll use 300 D vector representation of the words from pretrained embedding index 
# that we downloaded 

embedding_matrix=np.zeros((vocab_size+1,300)) #numpy matrix of size (2754,300).each row representing one word

for word,i in tk.word_index.items(): #word will become 'the' and index will become '1'
    embed_vector=embeding_index.get(word) #this would be glove vector for word 'the'
    if embed_vector is not None:
        embedding_matrix[i]=embed_vector #1st row of the embeding matrix is being updated by the glove vector word 'the'
        
# if there are specific words which are not present in pretrained embedding 
# their weights will remain 0. if there are too many such words 
# then you should probably not use pretrained embeddings 
```

```
embedding_matrix
```

```
inputs=Input(name='text_input',shape=[max_len]) #input will be 44 dim and 2d matrix

embed=Embedding(vocab_size+1,300,input_length=max_len,mask_zero=True,
                weights=[embedding_matrix],trainable=False)(inputs)  #300 is op we want #not to train the embending matrix because its already trained

lstm_layer=LSTM(50)(embed)

dense1=Dense(10,activation='relu')(lstm_layer)

drop=Dropout(0.2)(dense1)

final_layer=Dense(1,activation='sigmoid')(drop)

model=Model(inputs=inputs,outputs=final_layer)
```

```
model.summary()
```

```
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
model.fit(seq_train_matrix,y_train.values,validation_data=[seq_test_matrix,y_test.values],epochs=10,class_weight={0:1,1:12},
          batch_size=50)
```

```
p=model.predict(seq_test_matrix)
```

```
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,p)
```
